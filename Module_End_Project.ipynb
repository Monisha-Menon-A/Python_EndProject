```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r"C:\Users\Monisha Menon\Practice\myexcel.xlsx"
df = pd.read_excel(file_path, engine='openpyxl')
```

```python
# Preprocessing: Correct the height column
df['Height'] = np.random.randint(150, 181, size=len(df))
```

```python
# Analysis 1: Distribution of employees across teams
team_distribution = df['Team'].value_counts()
team_percentage = (team_distribution / team_distribution.sum()) * 100
```

```python
# Visualization 1: Bar chart for team distribution
plt.figure(figsize=(10, 6))
team_distribution.plot(kind='bar')
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()
```

The team with the highest number of employees is New Orleans Pelicans (19 employees).
The percentage distribution across teams shows that New Orleans Pelicans has the largest share at 4.15% of total employees.

```python
# Analysis 2: Segregation of employees by position
position_distribution = df['Position'].value_counts()
```

```python
# Visualization 2: Bar chart for position distribution
plt.figure(figsize=(10, 6))
position_distribution.plot(kind='bar')
plt.title('Segregation of Employees by Position')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()
```

SG and PF positions have the highest number of employees, while C have the lowest count.

```python
# Analysis 3: Predominant age group
df['Age Group'] = pd.cut(df['Age'], bins=[19, 25, 30, 35, 40])
age_group_distribution = df['Age Group'].value_counts().sort_index()
```

```python
# Visualization 3: Histogram for age group distribution
plt.figure(figsize=(10, 6))
age_group_distribution.plot(kind='bar')
plt.title('Predominant Age Group Among Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

The majority of employees fall within the 19-25 age range, followed by 25-30 years, suggesting that the employee set is relatively young. 
Employees above 35 years are very few in number.

```python
# Analysis 4: Team and position with highest salary expenditure
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().unstack()
```

```python
# Visualization 4: Heatmap for salary expenditure
plt.figure(figsize=(12, 8))
sns.heatmap(salary_expenditure, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Salary Expenditure by Team and Position')
plt.show()
```

Cleveland Cavaliers account for the highest salary expenditure among teams, and the C position has the highest salary allocation. 

```python
# Analysis 5: Correlation between age and salary
correlation = df['Age'].corr(df['Salary'])
```

```python
# Visualization 5: Scatter plot for age vs. salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title(f'Correlation Between Age and Salary (Correlation: {correlation:.2f})')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

There is a weak positive correlation (0.214) between age and salary, meaning that older employees tend to earn slightly more.
